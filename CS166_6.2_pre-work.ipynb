```python
import numpy as np
import scipy
from matplotlib import pyplot as plt
```

```python
# Exercise 14

rates = np.array([1, 2, 3, 4, 5])
betas = 1/rates

def average_waiting_time(beta):

    arrivals = []
    waitingtimes = []

    for i in range(100000):
        if i == 0:
            arrivals.append(scipy.random.exponential(beta))
        else:
            arrivals.append(scipy.random.exponential(beta) + arrivals[i - 1])

    for i in range(1, len(arrivals)):
        waitingtimes.append(arrivals[i] - arrivals[i - 1])
    
    print(f'Beta is {beta}, and average waiting time is {np.mean(waitingtimes)}')

for i in betas:
    average_waiting_time(i)
```

```python
# Exercise 24

def simulate():

    currentfund = 150
    age = 50
    alive = True
    
    while alive == True:
        
        if np.random.rand() <= data[age]:
            alive = False
            
        else:
            if age >= 70:
                currentfund = currentfund - 65
            else:
                currentfund = currentfund + scipy.random.normal(loc = currentfund*0.08, scale = currentfund*0.09, size = 1) + 10
            
            age += 1
            
    return currentfund

fundsattimeofdeath = []

for i in range(1000):
    fundsattimeofdeath.append(simulate())

plt.hist(fundsattimeofdeath, density = True)
plt.title('Funds at time of death')
plt.xlabel('Money in thousands of dollars')
plt.show()
```
