```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats as sts
```

```python
# Task 2. a) Reflection.

# Put a reflection barrier at x = –4, i.e., the walk may land at –4 but may not stay there, and
# the next step must be to –3. Plot a sample path and a histogram.

def random_walk_with_drift():

    time = np.arange(31)
    steps = [0]
    for i in range(1, len(time)):
        if np.random.rand() <= 0.5:
            x = steps[i - 1] - 1
        else:
            x = steps[i - 1] + 1
        if x == -4:
            steps.append(x)
            steps.append(-3)
        else:
            steps.append(x)
    time = np.arange(len(steps))
    return time, steps

time, steps = random_walk_with_drift()

print(f'The mean of 1 random walk is {np.mean(steps)}.')

plt.plot(time, steps, 'C0o')
plt.plot(time, steps, 'C0-')
plt.title('Sample path plot')
plt.xlabel('Time')
plt.ylabel('<- Right | Left ->')
plt.ylim(max(steps) + 1, min(steps) - 1)
plt.show()

plt.hist(steps)
plt.title('Histogram')
plt.show()
```

```python
# Task 2. b) Partial block.

# Next, add a partial block at x = 6, i.e., at x = 6 there is only a 0.25 chance of going to 7
# and a 0.75 chance of going to 5 (same for 7 to 6 ). Plot a histogram.

def random_walk_with_drift():

    time = np.arange(31)
    steps = [0]
    for i in range(1, len(time)):
        if np.random.rand() <= 0.5:
            x = steps[i - 1] - 1
        else:
            x = steps[i - 1] + 1
        if x == 6:
            steps.append(x)
            if np.random.rand() < 0.25:
                steps.append(7)
            else:
                steps.append(5)
        else:
            steps.append(x)
    time = np.arange(len(steps))
    return time, steps

time, steps = random_walk_with_drift()

print(f'The mean of 1 random walk is {np.mean(steps)}.')

plt.plot(time, steps, 'C0o')
plt.plot(time, steps, 'C0-')
plt.title('Sample path plot')
plt.xlabel('Time')
plt.ylabel('<- Right | Left ->')
plt.ylim(max(steps) + 1, min(steps) - 1)
plt.show()

plt.hist(steps)
plt.title('Histogram')
plt.show()
```
