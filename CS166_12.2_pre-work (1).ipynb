```python
def calculate_q(k):
    '''
    Use a numerical root finder to determine q from the equation
    q = exp(k*(qÂ­1)).
    '''
    from scipy.optimize import root
    return root(lambda q: q - np.exp(k * (q - 1)), 0).x[0]
```

```python
import numpy as np
import matplotlib.pyplot as plt

ks = np.linspace(1, 11)
solutions = np.array([calculate_q(i) for i in ks])
solutions = (1 - solutions) * 100

plt.plot(ks, solutions, color = 'green')
plt.title('Theoretical estimate for the size of the LCC')
plt.xlabel('k')
plt.ylabel('Size of the LCC (in the percentage of nodes)')
plt.show()
```
